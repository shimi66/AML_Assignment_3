Load the one star and five star parts of the data

```python
import gzip
import json
import csv
import pandas as pd
import gensim
from gensim.models import CoherenceModel
from gensim.parsing.preprocessing import STOPWORDS
from gensim.utils import simple_preprocess
from gensim import corpora
import multiprocessing
```

```python
#load in the compressed file to an iterator object

def parse(file_path):
    with gzip.open(file_path, 'r') as f:
        for compressed_json in f:
            obj = json.loads(compressed_json.decode('utf-8'))
            yield obj

path = 'amazon/Home_and_Kitchen.json.gz'
it = parse(path)

output_file1 = 'overall1.csv'
output_file2 = 'overall5.csv'


def extract_some(it, output_file1, output_file2):
    counter1 = 0
    counter2 = 0
    # Open the output file for writing
    with open(output_file1, mode='w', newline='') as file1, open(output_file2, mode='w', newline='') as file2:
        fieldnames = ['reviewText']
        writer1 = csv.DictWriter(file1,fieldnames=fieldnames)
        writer2 = csv.DictWriter(file2,fieldnames=fieldnames)
        writer1.writeheader()
        writer2.writeheader()
        for obj in it:
            if counter1 > 1_000_000 and counter2 > 1_000_000:
                break
            # With a 10% probability, write the 'asin' and 'reviewText' fields to the output file
            if 'reviewText' in obj and 'overall' in obj:
                if obj.get('overall') == 1:
                    if counter1 > 1_000_000:
                        continue
                    row = {'reviewText': obj.get('reviewText')}
                    writer1.writerow(row)
                    counter1+=1
                elif obj.get('overall') == 5:
                    if counter2 > 1_000_000:
                        continue
                    row = {'reviewText': obj.get('reviewText')}
                    writer2.writerow(row)
                    counter2+=1


extract_some(it, output_file1, output_file2)
```

Run LDA on one star

```python
# Define stopwords
stop_words = set(STOPWORDS)

output_file = output_file1

# Load data from CSV file
data = pd.read_csv(output_file, error_bad_lines=False)

# Defined stopwords from above

# Define a function to tokenize and preprocess the data
def preprocess(text):
    result = []
    for token in simple_preprocess(text):
        if token not in stop_words and len(token) > 3:
            result.append(token)
    return result

# Tokenize and preprocess the data
data_processed = data['reviewText'].astype(str).apply(preprocess)

# Create dictionary and corpus
dictionary = corpora.Dictionary(data_processed)
corpus = [dictionary.doc2bow(text) for text in data_processed]

# Train the LDA model
lda_model_rating1 = gensim.models.ldamulticore.LdaMulticore(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=10, 
                                                random_state=42,
                                                passes=10,
                                                per_word_topics=True,
                                                workers=multiprocessing.cpu_count())

# Print the topics

# Compute coherence score
cm = CoherenceModel(model=lda_model_rating1, texts=data_processed, dictionary=dictionary, coherence='u_mass')
c_1 = cm.get_coherence()
print('\nCoherence Score u_mass: ', c_1)
cm = CoherenceModel(model=lda_model_rating1, texts=data_processed, dictionary=dictionary, coherence='c_v')
c_2 = cm.get_coherence()
print('\nCoherence Score c_v: ', c_2)

#write the topics and coherences to a txt file
with open('1.txt', 'w') as f:
    for idx, topic in lda_model_rating1.print_topics(-1):
        f.write('Topic: {} \nWords: {}'.format(idx, topic))
        f.write('\nCoherence Score u_mass: ' + str(c_1))
        f.write('\nCoherence Score c_v: '+ str(c_2))
```

```python
#take the words in double quotes in 1.txt and 5.txt and read them in as a list of lists, with all of the words in double quotes on one line as one list
import re

# Define the paths to your input files
file1_path = '1.txt'
file5_path = '5.txt'

# Define a regular expression pattern to match double-quoted words
pattern = r'"([^"]*)"'

# Define a helper function to extract the double-quoted words from a line of text
def extract_quotes(line):
    return re.findall(pattern, line)

# Read in the double-quoted words from each file
with open(file1_path, 'r') as file1, open(file5_path, 'r') as file5:
    file1_quotes = [extract_quotes(line) for line in file1 if not line.startswith('Coherence')]
    file5_quotes = [extract_quotes(line) for line in file5 if not line.startswith('Coherence')]

#remove the lists that are empty
file1_quotes = [x for x in file1_quotes if x != []]
file5_quotes = [x for x in file5_quotes if x != []]
print(file1_quotes)
print("_____________________")
print(file5_quotes)
```
